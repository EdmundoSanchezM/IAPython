### Operacion aritmeticas

```python
import pandas as pd
```

```python
fruits = pd.Series(data = [20,40,60], index=['mango','apple','banana'])
fruits
```

```python
print(fruits + 3)
```

```python
print(fruits - 4)
```

```python
print(fruits * 2)
```

```python
print(fruits / 2)
```

```python
import numpy as np
```

```python
print(np.sqrt(fruits))
```

```python
print(np.exp(fruits))
```

```python
print(np.power(fruits,2))
```

```python
print(fruits['mango']+3)
```

```python
print(fruits[['banana','apple']] * 10)
```

```python
print(fruits.loc['apple'] / 10)
```
